Please input a list of integers, separated by commas, when prompted. The program will then process these integers to determine and display the mode score.

```python
def calculate_median(numbers):
    # Sort the list of numbers
    sorted_numbers = sorted(numbers)
    length = len(sorted_numbers)

    # Check if the length of the list is odd or even
    if length % 2 != 0:
        # If odd, return the middle element
        median = sorted_numbers[length // 2]
    else:
        # If even, calculate the average of the two middle elements
        median = (sorted_numbers[length // 2 - 1] + sorted_numbers[length // 2]) / 2

    return median

# Function to get user input and convert it into a list of integers
def get_user_input():
    user_input = input("Enter integers separated by commas: ")
    numbers = [int(x) for x in user_input.split(",")]
    return numbers

def main():
    numbers = get_user_input()
    median = calculate_median(numbers)
    print("Median:", median)

# Call the main function to execute the code
main()
```
