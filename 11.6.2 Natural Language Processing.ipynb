```python
!pip install textblob

from textblob import TextBlob

# Sample text for analysis
text = """
I had a fantastic day today! The weather was beautiful, and I enjoyed a lovely picnic with friends.
However, the traffic on the way home was terrible, and it took forever to get back.
Overall, I'm feeling pretty good about how everything went.
"""

# Create a TextBlob object
blob = TextBlob(text)

# Analyze the sentiment
sentiment = blob.sentiment

# Print the results with explanations
print(f"Polarity: {sentiment.polarity}")  # Polarity ranges from -1.0 (negative) to 1.0 (positive)
if sentiment.polarity > 0:
    print("Explanation: The text has an overall positive sentiment.")
elif sentiment.polarity < 0:
    print("Explanation: The text has an overall negative sentiment.")
else:
    print("Explanation: The text has a neutral sentiment.")

print(f"Subjectivity: {sentiment.subjectivity}")  # Subjectivity ranges from 0.0 (objective) to 1.0 (subjective)
if sentiment.subjectivity > 0.5:
    print("Explanation: The text is mostly subjective, indicating it expresses opinions or feelings.")
elif sentiment.subjectivity < 0.5:
    print("Explanation: The text is mostly objective, indicating it is based on facts.")
else:
    print("Explanation: The text is balanced between objectivity and subjectivity.")
```
