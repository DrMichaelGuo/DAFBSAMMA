```python
# Define a function to calculate the range score
def calculate_range_score():
    # Prompt the user for a list of integers separated by commas
    numbers_input = input("Please enter a list of integers, separated by commas: ")

    # Convert the input string to a list of integers
    numbers = list(map(int, numbers_input.split(',')))

    # Calculate the range score (maximum - minimum)
    range_score = max(numbers) - min(numbers)

    # Display the range score
    print(f"The range score is: {range_score}")

# Call the function to execute
calculate_range_score()
```
