```python
# Step 1: Import the necessary libraries
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Step 2: Load your dataset (Example: Assume 'df' is your DataFrame with features for clustering)
# Example dataset creation (you can replace this with your actual dataset)
data = {
    'feature1': np.random.rand(100),   # Random feature 1
    'feature2': np.random.rand(100),   # Random feature 2
    'feature3': np.random.rand(100)    # Random feature 3
}

# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)

# Step 3: Preprocess the data (optional but recommended)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Step 4: Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)  # Specify the number of clusters
kmeans.fit(scaled_data)  # Fit the K-Means model to the data

# Step 5: Add cluster labels to the original dataset
df['Cluster'] = kmeans.labels_

# Step 6: Visualise the clusters (for 2D data)
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=df['Cluster'], cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering')
plt.show()

# Step 7: View cluster centres
print("Cluster Centres:", kmeans.cluster_centers_)

# Step 8: Evaluate the K-Means model
print("Inertia (Sum of squared distances):", kmeans.inertia_)
```
