```python
import matplotlib.pyplot as plt
import numpy as np

# Let's create some sample data for hours studied and exam scores
# For the purpose of illustration, let's assume a positive correlation between them

np.random.seed(0)  # for reproducible results
hours_studied = np.random.uniform(1, 10, 100)  # Hours studied
exam_scores = 50 + 5 * hours_studied + np.random.normal(0, 10, 100)  # Exam scores

# Let's clip the exam scores to ensure the maximum score is 100
exam_scores = np.clip(exam_scores, None, 100)

# Now, let's create a scatter plot with the sample data

plt.figure(figsize=(12, 6))
plt.scatter(hours_studied, exam_scores, c='blue', alpha=0.5)
plt.title('Exam Scores vs Hours Studied')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Let's create some sample data for number of customer service calls and customer satisfaction scores
# Assuming an inverse relationship (negative correlation) between them

np.random.seed(0)  # for reproducible results
number_of_calls = np.random.randint(1, 11, 100)  # Number of customer service calls ranging from 1 to 10
customer_satisfaction = 11 - number_of_calls + np.random.normal(0, 1, 100)  # Satisfaction score on a scale of 0-10

# Now, let's create a scatter plot with the sample data

plt.figure(figsize=(12, 6))
plt.scatter(number_of_calls, customer_satisfaction, c='red', alpha=0.5)
plt.title('Customer Satisfaction vs Number of Service Calls')
plt.xlabel('Number of Customer Service Calls')
plt.ylabel('Customer Satisfaction Score')
plt.xlim(1, 10)  # Set x-axis to show numbers from 1 to 10
plt.ylim(0, 10)  # Set y-axis to show scores from 0 to 10
plt.xticks(np.arange(1, 11, 1))  # Ensure x-axis shows values from 1 to 10
plt.grid(True)
plt.show()
```

```python
# Let's create some sample data for number of employees and stock market performance of a company
# Assuming no relationship between them

np.random.seed(0)  # for reproducible results
number_of_employees = np.random.randint(50, 5000, 100)  # Number of employees
stock_market_performance = np.random.normal(0, 1, 100)  # Stock market performance (standardized around 0)

# Now, let's create a scatter plot with the sample data

plt.figure(figsize=(12, 6))
plt.scatter(number_of_employees, stock_market_performance, c='green', alpha=0.5)
plt.title('Stock Market Performance vs Number of Employees')
plt.xlabel('Number of Employees')
plt.ylabel('Stock Market Performance (Standardized)')
plt.grid(True)
plt.show()
```
