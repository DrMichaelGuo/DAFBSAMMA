```python
import matplotlib.pyplot as plt

# Sample data: ages of customers
customer_ages = [21, 22, 22, 23, 23, 23, 24, 25, 25, 25, 26, 27, 27, 28, 29, 29, 24, 30, 26, 24, 25, 30, 31, 32, 33, 34, 35, 36, 37]

# Creating the histogram
plt.figure(figsize=(15, 6))
plt.hist(customer_ages, bins=max(customer_ages)-min(customer_ages)+1, range=(min(customer_ages)-0.5, max(customer_ages)+0.5), color='skyblue', edgecolor='black', align='mid')

# Adding title and labels
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Setting x-axis ticks to display each individual age
plt.xticks(range(min(customer_ages), max(customer_ages)+1))

# Showing the chart
plt.show()
```
