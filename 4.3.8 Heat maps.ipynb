```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Create a DataFrame with the customer footfall data
footfall_data = {
    'London': [900, 850, 300, 1000, 1100, 1200, 700],
    'Manchester': [750, 700, 250, 850, 900, 950, 650],
    'Birmingham': [650, 600, 200, 750, 800, 850, 550],
    'Liverpool': [450, 400, 150, 500, 550, 600, 300]
}

days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
store_footfall = pd.DataFrame(footfall_data, index=days_of_week)

# Plotting the heatmap
plt.figure(figsize=(16, 9))
sns.heatmap(store_footfall, annot=True, cmap='coolwarm', fmt='d', linewidths=.5)

# Adding titles and labels
plt.title('Weekly Customer Footfall Heat Map')
plt.ylabel('Day of the Week')
plt.xlabel('City')
plt.show()
```
