```python
import matplotlib.pyplot as plt

# Sample data: Sales figures for product categories
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Beauty & Health', 'Sports & Outdoors']
sales = [95000, 75000, 55000, 40000, 65000]

# Define colors for each category
colors = ['skyblue', 'salmon', 'lightgreen', 'orchid', 'gold']

# Creating the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(categories, sales, color=colors)

# Adding title and labels
plt.title('Sales by Product Category')
plt.xlabel('Category')
plt.ylabel('Sales (Â£)')
plt.xticks(rotation=45)  # Rotate category names for better readability

# Adding sales figures on each bar
for bar, sale in zip(bars, sales):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(sale), ha='center', va='bottom')

# Adding grid
plt.grid(True)

# Showing the chart
plt.show()
```
