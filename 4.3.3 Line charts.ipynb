```python
import matplotlib.pyplot as plt
import calendar

# Sample data: Monthly sales figures over a year
months = range(1, 13)
sales = [142, 150, 157, 163, 172, 160, 178, 185, 193, 200, 210, 220]

# Define month names
month_names = [calendar.month_abbr[i] for i in months]

# Creating the line chart
plt.figure(figsize=(10, 6))
plt.plot(month_names, sales, marker='o', linestyle='-', color='b')

# Adding title and labels
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')

# Showing the chart
plt.show()
```
