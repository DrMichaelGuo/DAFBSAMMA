```python
# Step 1: Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler

# Step 2: Load your dataset
# Example dataset creation (you can replace this with your actual dataset)
data = {
    'feature1': np.random.rand(100),   # Random feature 1
    'feature2': np.random.rand(100),   # Random feature 2
    'feature3': np.random.rand(100)    # Random feature 3
}

# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)

# Step 3: Preprocess the data (optional but recommended)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Step 4: Apply Hierarchical Clustering
# Create the linkage matrix using the Ward method (minimizes the variance of clusters being merged)
Z = linkage(scaled_data, method='ward')

# Step 5: Plot the Dendrogram to visualise the clustering process
plt.figure(figsize=(10, 7))
dendrogram(Z, labels=df.index)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()

# Step 6: Optional: Truncate the dendrogram to show the desired number of clusters
plt.figure(figsize=(10, 7))
dendrogram(Z, labels=df.index, truncate_mode='level', p=4)  # Truncating to 4 levels of the hierarchy
plt.title('Truncated Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```
