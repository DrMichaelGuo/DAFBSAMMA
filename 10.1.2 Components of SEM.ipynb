```python
# Install graphviz if needed
# !pip install graphviz

from graphviz import Digraph

# Create a Digraph object for the SEM diagram
dot = Digraph(comment='SEM Path Diagram')

# Style adjustments
dot.attr(rankdir='LR')  # Left to Right layout
dot.attr('node', shape='box', fontsize='12', fontname='Arial', style='rounded')

# Latent variables (ellipses)
dot.node('CS', 'Customer Satisfaction', shape='ellipse', fontsize='14', fontname='Arial')
dot.node('CL', 'Customer Loyalty', shape='ellipse', fontsize='14', fontname='Arial')

# Observed variables (rectangles)
dot.node('CS1', 'CS1', fontsize='12', fontname='Arial')
dot.node('CS2', 'CS2', fontsize='12', fontname='Arial')
dot.node('CS3', 'CS3', fontsize='12', fontname='Arial')
dot.node('CL1', 'CL1', fontsize='12', fontname='Arial')
dot.node('CL2', 'CL2', fontsize='12', fontname='Arial')
dot.node('CL3', 'CL3', fontsize='12', fontname='Arial')

# Edges for measurement model (Latent to observed)
dot.edge('CS', 'CS1', label='1', fontsize='12', fontname='Arial')
dot.edge('CS', 'CS2', label='1.00', fontsize='12', fontname='Arial')
dot.edge('CS', 'CS3', label='0.22', fontsize='12', fontname='Arial')
dot.edge('CL', 'CL1', label='1', fontsize='12', fontname='Arial')
dot.edge('CL', 'CL2', label='1.26', fontsize='12', fontname='Arial')
dot.edge('CL', 'CL3', label='1.26', fontsize='12', fontname='Arial')

# Structural relationship (latent to latent)
dot.edge('CS', 'CL', label='0.52', fontsize='12', fontname='Arial')

# Render and save the diagram
output_path = '/mnt/data/sem_path_diagram_enhanced'
dot.render(output_path, format='png', cleanup=False)

# Display the diagram file
dot
```
