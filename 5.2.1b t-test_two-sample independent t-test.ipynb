```python
from scipy.stats import ttest_ind
import numpy as np

# Updated sample data for both groups
group1 = [39, 48, 52, 42, 29, 37, 32, 43, 44, 48, 46, 46, 31, 45, 48, 47, 36, 43, 37, 38, 24, 40, 41, 39, 43, 33, 45, 43, 57, 56]
group2 = [51, 53, 42, 34, 47, 51, 59, 59, 46, 47, 41, 38, 36, 65, 45, 46, 39, 56, 37, 48, 42, 53, 45, 40, 49, 53, 50, 52, 44, 47]

# Calculate mean and standard deviation for both groups
mean_group1 = np.mean(group1)
mean_group2 = np.mean(group2)
std_group1 = np.std(group1)
std_group2 = np.std(group2)

print("\nGroup 1:")
print(f"Mean: {mean_group1:.2f}")
print(f"Standard Deviation: {std_group1:.2f}\n")

print("Group 2:")
print(f"Mean: {mean_group2:.2f}")
print(f"Standard Deviation: {std_group2:.2f}\n")

# Performing the Two-Sample Independent t-test
t_stat, p_value = ttest_ind(group1, group2)

print(f"T-statistic: {t_stat:.2f}")
print(f"\nP-value: {p_value:.3f}")

# Interpreting the results
if p_value < 0.05:
    print("\nReject the null hypothesis.")
else:
    print("\nFail to reject the null hypothesis.")
```
