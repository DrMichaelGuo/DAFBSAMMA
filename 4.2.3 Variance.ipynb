Please input a list of integers, separated by commas, when prompted. The program will then process these integers to calculate and display the variance score.

```python
def calculate_variance(numbers):
    # Calculate the mean
    mean = sum(numbers) / len(numbers)

    # Calculate the squared deviations and variance
    squared_deviations = [(x - mean) ** 2 for x in numbers]
    variance = sum(squared_deviations) / len(numbers)

    return variance

user_input = input("Enter a list of integers, separated by commas: ")

try:
    numbers_list = [int(num) for num in user_input.split(',')]
    print(f"The variance is: {calculate_variance(numbers_list)}")
except ValueError:
    print("Invalid input. Please enter a list of integers separated by commas.")
```
