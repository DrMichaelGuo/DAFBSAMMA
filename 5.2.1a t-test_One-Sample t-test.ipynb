```python
from scipy.stats import ttest_1samp
import numpy as np

# Data Collection
sugar_intake = [53, 39, 45, 57, 54, 25, 45, 33, 34, 39, 36, 50, 43, 36, 39, 38, 50, 33, 38, 26, 9, 42, 44, 28, 58, 20, 35, 33, 50, 50, 37, 39, 26, 15, 32, 37, 47, 47, 31, 32, 25, 21, 18, 55, 30, 31, 22, 43, 19, 33]

# Hypotheses
population_mean = 30
# Null Hypothesis (H_0): The mean daily sugar intake is 30 grams.
# Alternative Hypothesis (H_1): The mean daily sugar intake is not 30 grams.

# Perform one-sample t-test using the sample data
t_stat, p_val = ttest_1samp(sugar_intake, population_mean)

print(f"\nSample Mean: {np.mean(sugar_intake):.2f} grams")
print(f"Sample Standard Deviation: {np.std(sugar_intake, ddof=1):.2f} grams")
print(f"T-statistic: {t_stat:.2f}")
print(f"\nP-value: {p_val:.4f}")

# Interpretation
alpha = 0.05
if p_val < alpha:
    print("\nReject the null hypothesis.")
else:
    print("\nFail to reject the null hypothesis.")
```
