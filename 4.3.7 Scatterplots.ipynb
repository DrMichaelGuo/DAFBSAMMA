```python
import matplotlib.pyplot as plt

# Sample data
advertising_spend = [200, 300, 400, 500, 600, 700, 800, 900, 1000, 1100, 1200, 1300]
sales_revenue = [600, 800, 950, 1100, 1200, 1300, 1500, 1650, 1800, 2000, 2100, 2200]

# Creating the scatterplot
plt.figure(figsize=(10, 6))
plt.scatter(advertising_spend, sales_revenue, color='blue')

# Adding title and labels
plt.title('Advertising Spend vs. Sales Revenue')
plt.xlabel('Advertising Spend')
plt.ylabel('Sales Revenue')

# Showing the chart
plt.show()
```
