Please input a list of integers, separated by commas, when prompted. The program will then process these integers to determine and display the range score.

```python
from collections import Counter
from scipy.stats import mode

# Function to calculate the mode
def calculate_mode(numbers):
    # Convert the list of strings to a list of integers
    numbers = [int(number.strip()) for number in numbers.split(',')]
    # Calculate the frequency of each number
    frequency = Counter(numbers)
    # Find the mode
    mode_value, count = mode(numbers)
    # Check if the dataset is unimodal, bimodal/multimodal or has no mode
    if len(frequency) == len(numbers):
        return "No mode", None
    else:
        # Find the highest frequency
        max_count = max(frequency.values())
        # Find all numbers that have the highest frequency
        modes = [number for number, freq in frequency.items() if freq == max_count]
        return "Unimodal" if len(modes) == 1 else "Bimodal or multimodal", modes

# Get input from the user
user_input = input("Enter a list of integers, separated by commas: ")

# Process the input to find the mode
modal_characteristic, mode_values = calculate_mode(user_input)

# Display the result
if modal_characteristic == "No mode":
    print("The dataset has no mode.")
else:
    print(f"The dataset is {modal_characteristic} with a mode of: {', '.join(map(str, mode_values))}")
```
