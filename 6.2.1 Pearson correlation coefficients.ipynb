# 1 Raw Scores Method

```python
import numpy as np

# Data
advertising_spend = np.array([20, 22, 25, 28, 30, 35])
sales_revenue = np.array([120, 150, 160, 180, 200, 220])

# Number of observations
n = len(advertising_spend)

# Sums of X, Y, X*Y, X^2, and Y^2
sum_X = np.sum(advertising_spend)
sum_Y = np.sum(sales_revenue)
sum_XY = np.sum(advertising_spend * sales_revenue)
sum_X2 = np.sum(advertising_spend**2)
sum_Y2 = np.sum(sales_revenue**2)

# Calculate the Pearson correlation coefficient using the raw scores method
numerator_raw = (n * sum_XY) - (sum_X * sum_Y)
denominator_raw = np.sqrt((n * sum_X2 - sum_X**2) * (n * sum_Y2 - sum_Y**2))

# Pearson correlation coefficient
pearson_correlation_raw = numerator_raw / denominator_raw

# Print the result
print(f"Pearson correlation using the Raw Scores Method Formula = {pearson_correlation_raw}")
```

# 2 The Covariance Method

```python
import numpy as np

# Data for advertising spend (in '000 pounds) and sales revenue (in '000 pounds)
advertising_spend = np.array([20, 22, 25, 28, 30, 35])
sales_revenue = np.array([120, 150, 160, 180, 200, 220])

# Calculate the means
mean_advertising_spend = np.mean(advertising_spend)
mean_sales_revenue = np.mean(sales_revenue)

# Calculate the numerator Σ(Xi - X̄)(Yi - Ȳ)
numerator = np.sum((advertising_spend - mean_advertising_spend) * (sales_revenue - mean_sales_revenue))

# Calculate the denominator √(Σ(Xi - X̄)²Σ(Yi - Ȳ)²)
denominator = np.sqrt(np.sum((advertising_spend - mean_advertising_spend)**2) * np.sum((sales_revenue - mean_sales_revenue)**2))

# Calculate the Pearson correlation coefficient using the covariance method
pearson_correlation_coefficient = numerator / denominator

# Print the result
print(f"Pearson correlation using the Covariance Method Formula = {pearson_correlation_coefficient}")

```
