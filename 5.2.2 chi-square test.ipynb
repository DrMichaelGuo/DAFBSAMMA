```python
import numpy as np
from scipy.stats import chi2

# Observed frequencies
observed = np.array([150, 200])

# Expected frequencies if no preference (assuming equal preference)
expected = np.array([175, 175])

# Calculate the chi-square statistic
chi_square_stat = ((observed - expected)**2 / expected).sum()

# Degrees of freedom
df = len(observed) - 1  # number of categories minus 1

# Calculate the p-value using the chi-square distribution
p_value = chi2.sf(chi_square_stat, df)

# Critical value for a significance level of 0.05
alpha = 0.05
critical_value = chi2.ppf(1 - alpha, df)

# Format chi-square statistic to two decimal places
chi_square_stat_formatted = "{:.2f}".format(chi_square_stat)

# Format p-value to three decimal places
p_value_formatted = "{:.3f}".format(p_value)

# Decision rule
if p_value < alpha:
    decision = "Reject the null hypothesis"
else:
    decision = "Fail to reject the null hypothesis"

print("Chi-square statistic =", chi_square_stat_formatted)
print("Degrees of freedom =", df)
print("Critical value for alpha = 0.05 and df =", df, "is", round(critical_value, 2))
print("p-value =", p_value_formatted)
print("Decision:", decision)
```
