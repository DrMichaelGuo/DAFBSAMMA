Please input a list of integers, separated by commas, when prompted. The program will then process these integers to determine and display the IQR score.

```python
def calculate_iqr(data):
    # Sort the data
    sorted_data = sorted(data)

    # Helper function to calculate median
    def median(dataset):
        n = len(dataset)
        if n % 2 == 0:
            median_value = (dataset[n//2 - 1] + dataset[n//2]) / 2
        else:
            median_value = dataset[n//2]
        return median_value

    # Find the medians for Q1 and Q3
    mid_index = len(sorted_data) // 2
    if len(sorted_data) % 2 == 0:
        lower_half = sorted_data[:mid_index]
        upper_half = sorted_data[mid_index:]
    else:
        lower_half = sorted_data[:mid_index]
        upper_half = sorted_data[mid_index+1:]

    q1 = median(lower_half)
    q3 = median(upper_half)

    # Calculate IQR
    iqr = q3 - q1
    return iqr

# Input from user
input_string = input("Please input a list of integers, separated by commas: ")
integer_list = [int(i) for i in input_string.split(",")]

# Calculate and display the IQR
iqr_score = calculate_iqr(integer_list)
print(f"The Interquartile Range (IQR) score is: {iqr_score}")
```
