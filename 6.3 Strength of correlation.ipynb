```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Function to generate data with a given correlation coefficient
def generate_correlated_data(r, n=100):
    mean = [0, 0]
    cov = [[1, r], [r, 1]]  # covariance matrix
    x, y = np.random.multivariate_normal(mean, cov, n).T
    return x, y

# Define correlation coefficients
correlations = [0, 0.21, 0.44, 0.85, 1.00]
titles = [
    "No Correlation (r = 0)",
    "Weak Correlation (r = 0.21)",
    "Moderate Correlation (r = 0.44)",
    "Strong Correlation (r = 0.85)",
    "Perfect Correlation (r = 1.00)"
]

# Define colors
colors = ["blue", "green", "orange", "red", "purple"]

# Set up the matplotlib figure
plt.figure(figsize=(15, 10))

# Generate and plot data for each correlation coefficient
for i, (r, title, color) in enumerate(zip(correlations, titles, colors)):
    x, y = generate_correlated_data(r)
    plt.subplot(2, 3, i + 1)
    sns.scatterplot(x=x, y=y, color=color)
    plt.title(title)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(True)

# Adjust layout
plt.tight_layout()

# Display the figure
plt.show()
```
