```python
import matplotlib.pyplot as plt

# Revised sales data including the 'East' region
sales_data = {
    'North': [120, 85, 90, 150, 130, 110, 115, 95, 145, 125],
    'South': [80, 70, 130, 90, 110, 75, 65, 105, 95, 85],
    'West': [105, 110, 75, 120, 115, 90, 130, 85, 100, 95],
    'East': [95, 105, 120, 100, 130, 110, 90, 115, 80, 105]
}

# Create a boxplot with four different colors
plt.figure(figsize=(10, 6))
boxprops = dict(linestyle='-', linewidth=2)
medianprops = dict(linestyle='-', linewidth=2, color='firebrick')

bp = plt.boxplot(sales_data.values(), patch_artist=True, boxprops=boxprops, medianprops=medianprops)
plt.xticks(range(1, len(sales_data) + 1), sales_data.keys())
plt.title('Sales Performance by Region')
plt.xlabel('Region')
plt.ylabel('Sales (in thousands)')

# Assigning different colors to each box
colors = ['lightblue', 'lightgreen', 'lightpink', 'lightyellow']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

plt.grid(True)
plt.show()
```
