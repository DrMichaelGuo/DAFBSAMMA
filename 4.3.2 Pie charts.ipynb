```python
import matplotlib.pyplot as plt

# Updated budget allocations
departments = ['R&D', 'Marketing', 'Operations', 'Sales', 'IT']
budget_allocations = [20000, 25000, 30000, 15000, 10000]

# Calculate total budget
total_budget = sum(budget_allocations)

# Creating the pie chart
plt.figure(figsize=(8, 8))
plt.pie(budget_allocations, labels=departments, autopct=lambda pct: f'£{pct/100*total_budget:.0f}\n({pct:.1f}%)', startangle=140, colors=['gold', 'lightgreen', 'lightcoral', 'lightskyblue', 'lavender'])

# Adding a title
plt.title('Annual Budget Allocation by Department')

# Showing the chart
plt.show()
```
