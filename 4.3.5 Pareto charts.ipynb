```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Provided data: Customer complaints and their frequencies
complaint_types = ['Payment Issues', 'Poor Customer Service', 'Late Delivery', 'Damaged Product', 'Wrong Item Sent']
frequencies = [45, 75, 150, 90, 40]

# Creating a DataFrame
data = pd.DataFrame({'Complaint Type': complaint_types, 'Frequency': frequencies})
data = data.sort_values('Frequency', ascending=False)
data['Cumulative'] = data['Frequency'].cumsum() / data['Frequency'].sum() * 100

# Creating the Pareto Chart
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(data['Complaint Type'], data['Frequency'], color='C0')

# Secondary axis for the cumulative percentage
ax2 = ax.twinx()
ax2.plot(data['Complaint Type'], data['Cumulative'], color='C1', marker='D', ms=7, linestyle='-', linewidth=2)
ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0f}%'))

# Adding titles and labels
plt.title('Pareto Chart of Customer Complaints')
ax.set_xlabel('Complaint Type')
ax.set_ylabel('Frequency')
ax2.set_ylabel('Cumulative Percentage (%)')

# Showing the chart
plt.show()
```
