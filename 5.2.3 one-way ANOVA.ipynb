```python
import scipy.stats as stats

# Given data
n = 20  # number of samples for each programme

# Programme A data
mean_A = 50  # mean sales in thousands of pounds
var_A = 5  # variance in sales in thousands of pounds

# Programme B data
mean_B = 55
var_B = 8

# Programme C data
mean_C = 60
var_C = 6

# Calculate the sum of squares for each programme (SS_within)
SS_within_A = (n - 1) * var_A
SS_within_B = (n - 1) * var_B
SS_within_C = (n - 1) * var_C

# Sum of SS_within for all programmes
SS_within = SS_within_A + SS_within_B + SS_within_C

# Calculate the degrees of freedom within (df_within)
df_within = n * 3 - 3

# Calculate the mean square within (MS_within)
MS_within = SS_within / df_within

# Grand mean
grand_mean = (mean_A + mean_B + mean_C) / 3

# Calculate the sum of squares between (SS_between)
SS_between = n * ((mean_A - grand_mean)**2 + (mean_B - grand_mean)**2 + (mean_C - grand_mean)**2)

# Calculate the degrees of freedom between (df_between)
df_between = 3 - 1

# Calculate the mean square between (MS_between)
MS_between = SS_between / df_between

# Calculate the F-statistic
F_statistic = MS_between / MS_within

# Calculate the p-value using the F-distribution
p_value = stats.f.sf(F_statistic, df_between, df_within)

# Output the F-statistic, p-value, and degrees of freedom
print("F =", F_statistic)
print("p-value =", p_value)
print("Degrees of freedom (between) =", df_between)
print("Degrees of freedom (within) =", df_within)
```

```python
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Given data
n = 20  # number of samples for each programme

# Programme A data
mean_A = 50  # mean sales in thousands of pounds
var_A = 5  # variance in sales in thousands of pounds

# Programme B data
mean_B = 55
var_B = 8

# Programme C data
mean_C = 60
var_C = 6

# Combine data
sales_data = np.concatenate([np.random.normal(mean_A, np.sqrt(var_A), n),
                             np.random.normal(mean_B, np.sqrt(var_B), n),
                             np.random.normal(mean_C, np.sqrt(var_C), n)])

# Create group labels
labels = ['A'] * n + ['B'] * n + ['C'] * n

# Perform Tukey's HSD test
tukey_results = pairwise_tukeyhsd(sales_data, labels)

# Print the summary of the test results
print(tukey_results)
```
