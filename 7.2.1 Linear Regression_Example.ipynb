```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Generate a dataset
# Set a random seed for reproducibility
np.random.seed(0)
x = np.random.rand(100) * 100  # Random values for advertising spend
y = 2.5 * x + np.random.normal(size=x.size) * 20  # Sales revenue with some noise

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Create a scatter plot
plt.figure(figsize=(16, 9))
plt.scatter(x, y, color='blue', alpha=0.5, label='Data points')
plt.title('Scatter Plot with Regression Line')
plt.xlabel('X')
plt.ylabel('Y')

# Add the fitted regression line
plt.plot(x, slope * x + intercept, color='red', label=f'Fitted line: y = {slope:.2f}x + {intercept:.2f}')
plt.legend()

# Show the plot
plt.show()


```
