Please input a list of integers, separated by commas, when prompted. The program will then process these integers to calculate and display the median score.

```python
def calculate_mean(integers):
    return sum(integers) // len(integers)

def main():
    # Prompt the user for input
    input_string = input("Enter integers separated by commas: ")

    # Split the string into a list of strings on commas
    integer_strings = input_string.split(',')

    # Convert list of strings to a list of integers
    integers = [int(n.strip()) for n in integer_strings]

    # Calculate the mean
    mean = calculate_mean(integers)

    # Print the result
    print(f"The mean score is: {mean}")

# Call the main function to execute the code
main()
```
